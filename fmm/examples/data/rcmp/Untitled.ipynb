```python
import json
import numpy as np
import matplotlib
# matplotlib.use('QtAgg') # Replace 'other_backend' with a backend like 'Agg', 'TkAgg', etc.

import matplotlib.pyplot as plt
import tikzplotlib
```

```python
with open('double/double.json') as f:
    double = json.load(f)
    
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
data = [
    double['3']['ranks'],
    double['6']['ranks'],
    double['9']['ranks'],
]


# for (d, ax) in list(zip(data, [ax1])):
min_bin = min(data[0]) - 0.5
max_bin = max(data[2]) + 1.5
bins = np.arange(min_bin, max_bin)
    
# Create histogram
ax.hist(data, bins=bins, align='mid', label=['$\epsilon = 10^{-4}$', '$\epsilon = 10^{-7}$', '$\epsilon = 10^{-10}$'])

# Set xticks to integers
# int_ticks = np.arange(min(data[0]), max(data[2]) + 1)
# ax.set_xticks(int_ticks)
ax.set_ylabel('Frequency')
ax.set_xlabel('Rank $k$')
ax.legend()
plt.tight_layout()
# plt.legend(loc='upper left', bbox_to_anchor=(1.1, 1))
tikzplotlib.save("rcmp_double_precision.tex")
plt.show()
```
