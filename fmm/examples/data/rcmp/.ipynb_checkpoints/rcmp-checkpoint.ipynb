```python
import json
import numpy as np
import matplotlib.pyplot as plt
import tikzplotlib
```

```python
with open('double/double.json') as f:
    double = json.load(f)

with open('single/single.json') as f:
    single = json.load(f)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 4), sharex=True)

data = [
    single['3']['k_t'],
    single['5']['k_t'],
]

# for (d, ax) in list(zip(data, [ax1])):
min_bin = min(data[0]) - 0.5
max_bin = max(data[1]) + 1.5
bins = np.arange(min_bin, max_bin)
ax1.hist(data, bins=bins, align='mid', label=['$\epsilon = 10^{-3}$, $k=10$', '$\epsilon = 10^{-5}, k=47$'])

data = [
    double['6']['k_t'],
    double['8']['k_t'],
    double['10']['k_t'],
]
# for (d, ax) in list(zip(data, [ax1])):
min_bin = min(data[0]) - 0.5
max_bin = max(data[2]) + 1.5
bins = np.arange(min_bin, max_bin)
ax2.hist(data, bins=bins, align='mid', label=['$\epsilon = 10^{-6}$, $k=68$', '$\epsilon = 10^{-8}, k=133$', '$\epsilon = 10^{-10}, k=226$'])

# ax1.set_ylabel('Frequency')
# ax2.set_ylabel('Frequency')
ax2.set_xlabel('Rank $k_t$')
ax1.set_title("Single Precision")
ax2.set_title("Double Precision")
ax1.legend()
ax2.legend()
plt.tight_layout()
fig.text(-0.01, 0.5, 'Frequency', va='center', rotation='vertical')
tikzplotlib.save("rcmp.tex") 
plt.show()
```
