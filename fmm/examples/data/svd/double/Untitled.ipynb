```python
import json
import numpy as np
import matplotlib
# matplotlib.use('QtAgg') # Replace 'other_backend' with a backend like 'Agg', 'TkAgg', etc.

import matplotlib.pyplot as plt
import tikzplotlib
```

```python
with open('double.json') as f:
    double = json.load(f)
    
with open('../single/single.json') as f:
    single = json.load(f)
    
# Normalise singular values

for (k, v) in double.items():
    v['sigma'] = np.array(v['sigma'])/max(v['sigma'])
    
for (k, v) in single.items():
    v['sigma'] = np.array(v['sigma'])/max(v['sigma'])
    
keys = sorted([int(k) for k in double.keys()])

tmp = {}
for k in keys:
    tmp[k] = double[str(k)]
    
double = tmp

# keys = sorted)))

tmp = {}
for k in keys:
    tmp[k] = single[str(k)]
    
single = tmp
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4), sharey=True)

ax1.set_yscale('log')
ax2.set_yscale('log')
# for label in ax.get_xticklabels():
#     label.set_fontproperties(cmu_font)
# for label in ax.get_yticklabels():
#     label.set_fontproperties(cmu_font)
    
# line_styles = ['-', '--', '-.', ':', 
#                '-', '--', '-.', ':', '-']
for (i, (k, v)) in enumerate(double.items()):
    ax1.plot(v['sigma'], label=f'p={k}')

for (i, (k, v)) in enumerate(single.items()):
    ax2.plot(v['sigma'], label=f'p={k}')


# arial_unicode_font = {'fontname': 'Arial Unicode'}



# plt.legend()
plt.legend(loc='upper left', bbox_to_anchor=(1.1, 1))
tikzplotlib.save("laplace_singular_values.tex")

plt.show()

```
