```python
import json
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

import tikzplotlib
```

```python
with open('double/double.json') as f:
    double = json.load(f)
    
with open('single/single.json') as f:
    single = json.load(f)
    
# Normalise singular values

for (k, v) in double.items():
    v['sigma'] = np.array(v['sigma'])/max(v['sigma'])
    
for (k, v) in single.items():
    v['sigma'] = np.array(v['sigma'])/max(v['sigma'])
    
keys = sorted([int(k) for k in double.keys()])

tmp = {}
for k in keys:
    tmp[k] = double[str(k)]
    
double = tmp

keys = sorted([int(k) for k in single.keys()])

tmp = {}
for k in keys:
    tmp[k] = single[str(k)]
    
single = tmp
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4), sharey=True)

ax1.set_yscale('log')
ax2.set_yscale('log')

for (i, (k, v)) in enumerate(double.items()):
    ax2.plot(v['sigma'], label=f'p={k}')

for (i, (k, v)) in enumerate(single.items()):
    ax1.plot(v['sigma'], label=f'p={k}')

# ax1.set_xlim(0, 175)
# ax2.set_xlim(380, 387)

ax2.set_xlabel('Double Precision')
ax1.set_xlabel('Single Precision')

plt.legend(loc='upper left', bbox_to_anchor=(1.1, 1))
# tikzplotlib.save("laplace_singular_values.tex")

plt.show()

```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10, 3), sharey=True)

continuous = double[11]['sigma']

six = double[6]['sigma']
six_log_error = np.log(continuous[:len(six)]) - np.log(six)

eight = double[8]['sigma']
eight_log_error = np.log(continuous[:len(eight)]) - np.log(eight)

ten = double[10]['sigma']
ten_log_error = np.log(continuous[:len(ten)]) - np.log(ten)

ax1.plot(six_log_error)
ax2.plot(eight_log_error)
ax3.plot(ten_log_error)
ax3.set_xlim(0, 500)

ax1.set_ylabel('$\log(\epsilon)$')

fig.suptitle('Double Precision', fontsize=16)

ax1.text(0.1, 0.9, '$p=6$', fontsize=12,  va='center', transform=ax1.transAxes)
ax2.text(0.1, 0.9, '$p=8$', fontsize=12,  va='center', transform=ax2.transAxes)
ax3.text(0.1, 0.9, '$p=10$', fontsize=12,  va='center', transform=ax3.transAxes)

ax1.set_xlabel('$N$')
ax2.set_xlabel('$N$')
ax3.set_xlabel('$N$')
tikzplotlib.save("laplace_singular_values_error_double.tex")

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(6.66, 3), sharey=True)

continuous = single[11]['sigma']

three = single[3]['sigma']
three_log_error = np.log(continuous[:len(three)]) - np.log(three)

five = single[5]['sigma']
five_log_error = np.log(continuous[:len(five)]) - np.log(five)

ax1.plot(three_log_error)
ax2.plot(five_log_error)
ax1.text(0.17, 0.9, '$p=3$', fontsize=12, ha='center', va='center', transform=ax1.transAxes)
ax2.text(0.1, 0.9, '$p=5$', fontsize=12,  va='center', transform=ax2.transAxes)
ax1.set_ylabel('$\log(\epsilon)$')
ax1.set_xlabel('$N$')
ax2.set_xlabel('$N$')

fig.suptitle('Single Precision', fontsize=16)

tikzplotlib.save("laplace_singular_values_error_single.tex")

plt.show()
```
