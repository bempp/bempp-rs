```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Convergence

```python
pvec = np.array([2, 3, 4, 5, 6, 2, 3, 4, 5, 6, 2, 3, 4, 5, 6, 2, 3, 4, 5, 6])
nvec = np.array([1000000]*20)
kvec = np.array([10, 10, 10, 10, 10, 50, 50, 50, 50, 50, 150, 150, 150, 150, 150, np.inf, np.inf, np.inf, np.inf, np.inf])
rel_err_vec = np.array(
    [
        0.0396987004528946, 0.0005776929889951893, 9.092219386931128e-5, 0.0004069152829229744,  0.00010450533633955941,
        0.028480908738693186, 0.0004438004393188409, 6.567236243443253e-5, 2.600395188455199e-6, 1.0305288374677816e-6,
        0.03656674500641478, 0.00047444738728249434, 0.00017140577709560817, 3.448712731606661e-6, 4.2567160373434735e-7,
        0.03494759254752179, 0.0004453248536567167, 4.952565373626558e-5, 3.6011059379417663e-6, 3.4473758136713903e-7
    ]
)

```

```python
df = pd.DataFrame.from_dict({"p": pvec, "n": nvec, "k": kvec, "err": rel_err_vec})
```

```python
df
```

```python
fig, ax = plt.subplots()

ax.plot(df[df['k'] == 10]['p'], df[df['k'] == 10]['err'], '-o', label='10')
ax.plot(df[df['k'] == 50]['p'], df[df['k'] == 50]['err'], '-o', label='50')
ax.plot(df[df['k'] == 150]['p'], df[df['k'] == 150]['err'], '-o', label='150')
ax.plot(df[df['k'] == np.inf]['p'], df[df['k'] == np.inf]['err'], '-o', label='uncompressed')


ax.set_title('NPoints = 1e6, NCrit = 150, Adaptive Tree (~30k boxes)')
ax.set_xlabel('Expansion Order $p$')
ax.set_ylabel('Relative Error')
ax.set_yscale("log")
ax.set_xticks(pvec)
plt.legend(title='M2L SVD Compression Rank')
plt.show()
```
